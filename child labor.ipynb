```r
# install.packages("tidyverse")

# Load the tidyverse package (similar to importing libraries in Python)
library(tidyverse)
```

```r
# Load the dataset (`pd.read_csv` equivalent)

df <- "population of children by age, region, and sex.csv" 

population <- read_csv(df, skip = 1, na = c("..", "", "NA"))
```

```r
# Structure of the data frame (`df.info()` equivalent)

str(population)
```

```r
# First few rows of the data frame (`df.head()` equivalent)

head(population)
```

```r
# Column names 

colnames(population)
```

```r
# Clean and unpivot the data (similar to `pd.melt` in Python)

population_long <- population %>%
  pivot_longer(
    cols = -c(Age),
    names_to = "Age_Group",
    values_to = "Population"
  ) %>%
  mutate(Population = as.numeric(Population)) %>%
  drop_na(Population)

  population_long
```
